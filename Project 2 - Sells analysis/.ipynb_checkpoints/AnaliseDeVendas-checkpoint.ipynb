### Qual cliente comprou mais vezes?

```python
frequencia_cliente = vendas_df['Email cliente'].value_counts()
display(frequencia_cliente)
```

```python
frequencia_cliente.plot(figsize=(15, 5), yticks=range(0, 100, 5))
```

```python
#Escolhendo apenas os 5 primeiros clientes
frequencia_cliente[:5].plot(figsize=(15, 5), yticks=range(0, 100, 5))
```

### Qual a loja que mais vendeu?

```python
vendas_lojas = vendas_df.groupby('Nome da Loja').sum()
vendas_lojas
```

```python
vendas_lojas = vendas_lojas.sort_values('Quantidade Vendida', ascending = False)
display(vendas_lojas)
```

```python
vendas_lojas = vendas_lojas[['Quantidade Vendida']]
vendas_lojas.reset_index()
```

```python
vendas_lojas[:5].plot(figsize=(15, 5), yticks=range(0, 1500000, 100000), kind='bar')
```

```python
maior_vendas = vendas_lojas['Quantidade Vendida'].max()
loja_mais_vendeu = vendas_lojas['Quantidade Vendida'].idxmax()
print(loja_mais_vendeu, maior_vendas)
```

```python
menor_vendas = vendas_lojas['Quantidade Vendida'].min()
loja_menos_vendeu = vendas_lojas['Quantidade Vendida'].idxmin()
print(loja_menos_vendeu, menor_vendas)
```

### Alterando o formato da coluna Data da Venda e Data do Envio

```python
vendas_df.info()
```

```python
vendas_df['Data da Venda'] = pd.to_datetime(vendas_df['Data da Venda'], format='%d/%m/%YY')
vendas_df['Data do Envio'] = pd.to_datetime(vendas_df['Data do Envio'], format='%d/%m/%YY')
display(vendas_df)
```

### Criando uma nova coluna com as datas das vendas

```python
vendas_df["Dia da Venda"] = vendas_df["Data da Venda"].dt.day
vendas_df["Mes da Venda"] = vendas_df["Data da Venda"].dt.month
vendas_df["Ano da Venda"] = vendas_df["Data da Venda"].dt.year
display(vendas_df)
```
